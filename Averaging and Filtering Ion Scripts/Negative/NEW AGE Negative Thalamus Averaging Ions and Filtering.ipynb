```python
number_pixels_list_left_thalamus = []

for file in glob.glob('D:/NEW All AGE Negative masks and roi sums/Left Thalamus/*pixel_matrix_with_mask_left_thalamus.csv'):
    print(file)
    brain_matrix_without_background = pd.read_csv(file, index_col = 0)
    number_pixels_list_left_thalamus.append(len(brain_matrix_without_background[brain_matrix_without_background.sum(axis = 1) != 0]))
    
number_pixels_list_right_thalamus = []

for file in glob.glob('D:/NEW All AGE Negative masks and roi sums/Right Thalamus/*pixel_matrix_with_mask_right_thalamus.csv'):
    print(file)
    brain_matrix_without_background = pd.read_csv(file, index_col = 0)
    number_pixels_list_right_thalamus.append(len(brain_matrix_without_background[brain_matrix_without_background.sum(axis = 1) != 0]))
```
