Below we find the number of pixels in each brain so we can normalize each sum by the brain size.

```python
number_pixels_list_left_hippocampus = []

for file in glob.glob('Z:/All AGE Positive masks and roi sums/Left Hippocampus/*pixel_matrix_with_mask_left_hippocampus.csv'):
    print(file)
    brain_matrix_without_background = pd.read_csv(file, index_col = 0)
    number_pixels_list_left_hippocampus.append(len(brain_matrix_without_background[brain_matrix_without_background.sum(axis = 1) != 0]))
    
number_pixels_list_right_hippocampus = []

for file in glob.glob('Z:/All AGE Positive masks and roi sums/Right Hippocampus/*pixel_matrix_with_mask_right_hippocampus.csv'):
    print(file)
    brain_matrix_without_background = pd.read_csv(file, index_col = 0)
    number_pixels_list_right_hippocampus.append(len(brain_matrix_without_background[brain_matrix_without_background.sum(axis = 1) != 0]))
```
