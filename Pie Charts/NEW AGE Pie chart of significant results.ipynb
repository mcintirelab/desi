```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

labels = ['Cortex [4]', 'Globus Pallidus [19]', 'Hippocampus [15]', 'Hypothalamus [6]', 'Striatum [16]', 'Thalamus [74]', 'White Matter [22]', 'Whole Brains [20]']
sizes = [4, 19, 15, 6, 16, 74, 22, 20]
colors = ['red', 'yellow', 'green', 'blue', 'pink', 'orange', 'purple', 'brown']
dataframe_anova = pd.DataFrame([labels, sizes, colors])
dataframe_anova_transposed = dataframe_anova.T
dataframe_anova_transposed_to_keep = dataframe_anova_transposed[dataframe_anova_transposed[1] > 0]
dataframe_anova_transposed_to_keep_sorted = dataframe_anova_transposed_to_keep.sort_values(by = 1)
fig, ax = plt.subplots()
plt.title('ANOVA WT Significant Ion Distribution')
ax.pie(dataframe_anova_transposed_to_keep_sorted[1], labels=dataframe_anova_transposed_to_keep_sorted[0], colors = dataframe_anova_transposed_to_keep_sorted[2])
plt.savefig('Images For Paper/NEW_AGE_positive_and_negative_wt_only_significant_ion_pie_chart.tiff')
```
